### DROP CAR NAME

```python
df.drop('car_name', axis=1)
```

### NARROW DATA DOWN TO FEATURES TO FOCUS ON

```python
df1 = df[['displacement', 'acceleration', 'cylinders']]
df1
```

### NARROW DOWN FURTHER TO JUST TWO FEATURES

```python
x = df1.drop('cylinders', axis=1)
y = df1['cylinders']
x.head()
```

### GET X AND Y AXIS SHAPE

```python
x.shape, y.shape
```

### TRAIN TEST SPLIT
### Use train_test_split to split your data into Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

### BUILDING KNN MODEL INSTANCE WITH N_NEIGHBORS = 2 AND FITTING IT TO THE TRAIN DATA

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train, y_train)
```

### CHECK FOR THE ACCURACY OF YOUR MODEL

```python
from sklearn.metrics import accuracy_score

pred = knn.predict(x_test)
print('Model Accuracy = ', accuracy_score(pred, y_test))
```

### PREDICT AND EVALUATE

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```

### CHOOSING A K VALUE

```python
error_rate = []
for i in range (1, 40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    pred_i = knn.predict(x_test)
    error_rate.append(np.mean(pred_i != y_test))
```

### CREATING A PLOT USING INFO FROM THE FOR LOOP

```python
plt.figure(figsize=(16,8))
plt.plot(range(1, 40), error_rate, color='green', linestyle='dashed', marker='o', markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Errot Rate')
```

### RETRAIN WITH NEW K VALUE

```python
knn = KNeighborsClassifier(n_neighbors=30)

knn.fit(x_train, y_train)
pred = knn.predict(x_test)

print('WITH K=30')
print('\n')
print(confusion_matrix(y_test, pred))
print('\n')
print(classification_report(y_test, pred))
```
